In the context of data or analysis, the terms "high bound" and "low bound" refer to the upper and lower limits of a range. These bounds are used to establish the maximum and minimum possible values within that range.
The high bound represents the upper limit or maximum value that a variable or measurement can reach within a given range. It provides an upper boundary or ceiling beyond which the variable cannot go.

As explained, majority of the columns with the name "high bound" and "low bound" are estimations and I will not need them in this data visualization.
